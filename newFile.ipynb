```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# Set visualisation colours
mycols = ["#66c2ff", "#5cd6d6", "#00cc99", "#85e085", "#ffd966", "#ffb366", "#ffb3b3", "#dab3ff", "#c2c2d6"]
sns.set_palette(palette = mycols, n_colors = 4)
print('My colours are ready! :)')
train_df = pd.read_csv('train_dummified.csv')
test_df = pd.read_csv('test_dummified.csv')
y_read = pd.read_csv('price.csv')
y = np.log(y_read['SalePrice'])
y_ad = np.log(y_read['SalePriceAd'])
X = train_df
X_test=test_df
X.head()
```

```python
##params
lasso = Lasso()
alphas = np.linspace(0.0000,0.002,50)
lasso.set_params(normalize=True)
coefs_lasso  = []
scores = []
for alpha in alphas:
        lasso.set_params(alpha=alpha)
        lasso.fit(X, y)  
        coefs_lasso.append(lasso.coef_)
        scores.append(lasso.score(X, y))
        
coefs_lasso = pd.DataFrame(coefs_lasso,index=alphas, columns = X.columns)  
coefs_lasso.head()
```

```python
def numberToWords(num):
    single_digits = ["Zero", "One", "Two", "Three",  
                 "Four", "Five", "Six", "Seven",  
                 "Eight", "Nine"];
    ten_digits = {10: "Ten", 11: "Eleven", 12: "Twelve",  
              13: "Thirteen", 14: "Fourteen", 15: "Fifteen",  
              16: "Sixteen", 17: "Seventeen", 18: "Eighteen", 
              19: "Nineteen"}; 
    double_digits = {'2': "Twenty", '3': "Thirty", '4': "Forty", 
                 '5': "Fifty", '6': "Sixty", '7': "Seventy", '8': "Sighty",  
                 '9': "Ninety"};
    more_digits = ["Hundred", "Thousand", "Million", "Billion"]; 

    second = num % 10
    first = num // 10
    word = []
    index = 0
    hundred = 0
    while True:

        if index % 4 == 1:
            if second <= 19:
                word.append(ten_digits[second])
            elif second >= 20:
                firstdigit = str(second)[0]
                seconddigit = str(second)[1]
                word.append(single_digits[int(seconddigit)])
                word.append(double_digits[firstdigit])
        elif index % 4 == 2:
            firstdigit = str(second)[0]
            word.append(more_digits[hundred])
            word.append(single_digits[int(firstdigit)])

            hundred += 1
        elif index % 4 == 3:
            firstdigit = str(second)[0]
            word.append(more_digits[hundred])
            word.append(single_digits[int(firstdigit)])
            
            hundred += 1
        if first == 0:
            break
        index += 1
        second = first % 10 * 10**(index) + second
        first = first // 10
        
            
        print(first, second)
        
    print(word)    
    return ' '.join(word[::-1])

numberToWords(3111)
```
